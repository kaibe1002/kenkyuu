```python
class ML_loss(nn.Module):
  def __inif__(self):
    super().__init__()

  def forward(self,outputs,targets):
    sum=0
    for i in range(len(outputs)):
      sum=sum+(targets[i]*(torch.log(outputs[i,0]))-(outputs[i,0]))   

    loss=(-1)*sum
    return loss
```

```python
def rmse2(outputs,y):
  sum=0
  for i in range(len(outputs)):
    sum=sum+(outputs[i,0]-y.iloc[i])**2
    
  mse=sum/len(outputs)
  mse=mse.clone().detach().numpy()
  rmse=np.sqrt(mse)
  return rmse
```
